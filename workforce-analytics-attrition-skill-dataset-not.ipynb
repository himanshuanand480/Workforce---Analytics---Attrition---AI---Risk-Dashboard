## 🎯 Business Problem

Organizations face high employee attrition, which leads to increased hiring costs and reduced productivity.

This project aims to identify key factors influencing employee attrition and provide actionable insights for better workforce management.


**DATASET OVERVIEW:-**

📁 Dataset OverviewThe dataset includes employee details such as age, department, salary, job satisfaction, and attrition status, along with job-related skill requirements.It is designed to simulate real-world workforce scenarios.


```python
import numpy as np 
import pandas as pd 
```

**LOAD DATA**

```python
import pandas as pd

emp = pd.read_csv('/kaggle/input/datasets/himanshuanand480/workforce-analytics-attrition-and-skill-dataset/employee_data.csv')
job = pd.read_csv('/kaggle/input/datasets/himanshuanand480/workforce-analytics-attrition-and-skill-dataset/job_data.csv')

emp.head()
```

**📊 Analysis & Visualizations**

**📊 Attrition Rate**

```python
# Attrition rate
(emp['Attrition'] == 'Yes').mean() * 100
```

**📊 Department Analysis**

```python
import matplotlib.pyplot as plt

emp.groupby('Domain')['Attrition'].value_counts().unstack().plot(kind='bar')
plt.title("Attrition by Department")
plt.show()
```

**📊 Salary vs Attrition**

```python
emp.groupby('Attrition')['Salary '].mean()
```

```python
emp.columns
```

**📊 Work-Life Balance**

```python
emp.groupby('Work_Life_Balance')['Attrition'].value_counts().unstack().plot(kind='bar')
plt.show()
```

**📊 Skill Demand**

```python
job['Required_Skill'].value_counts().plot(kind='bar')
plt.show()
```

## 💡 Key Insights

- Employees with lower job satisfaction show higher attrition  
- Poor work-life balance leads to higher employee turnover  
- Salary impacts employee retention  
- Certain skills are in higher demand  

## 💼 Business Recommendations

- Improve employee engagement and satisfaction  
- Offer better work-life balance policies  
- Provide competitive salaries  
- Focus hiring on high-demand skills
